```python
if False:
    testset = pd.read_csv("tmp/test_set_stg.csv")
    testset = testset.dropna()
    print(testset.columns)
    print(len(testset))

    from inference import SpikeFeatureExtractor

    sfe = SpikeFeatureExtractor(truncate_burst=True)
    # Extract features for the entire DataFrame
    r = sfe.extract_from_dataframe(testset,True, 19)
    # we put back the g_s, g_u, label_p columns
    r['g_s'] = testset['g_s']
    r['g_u'] = testset['g_u']

    # drop nan
    r = r.dropna()
    # drop label == 0
    r = r[r['label'] != 0]

    print(r.columns, len(r))

    # group by ('g_u', 'g_s')
    # the first instance in each group is labeled 'picked', the rest are labeled 'not picked'
    r['label_p'] = r.groupby(['g_u', 'g_s']).cumcount().apply(lambda x: 'picked' if x == 0 else 'not picked')

    # Check the updated DataFrame
    print(r.head())

    # Step: Keep only 100 unique (g_u, g_s) pairs
    unique_pairs = r[['g_u', 'g_s']].drop_duplicates().head(len(r))
    r = r.merge(unique_pairs, on=['g_u', 'g_s'])

    print(f"Filtered to {len(r)} rows from 100 (g_u, g_s) groups.")

    from tqdm import tqdm
    from multiprocessing import Pool, cpu_count

    # we compute the mean the following : 

    r"""\paragraph{Evaluation metrics} During evaluation, two primary metrics are used. The loss $\mathcal{L}_\text{DICs}$ is employed to assess whether the model is learning correctly and is also used for hyperparameter tuning on the validation set $\mathcal{T}_\text{val}$. For the final evaluation on the test set $\mathcal{T}_\text{test}$, an additional performance measure is introduced. The reasoning is that interpreting the results of DICs regression directly is challenging, as it is difficult to translate the regression error into a meaningful interpretation of the quality of the generated populations. Therefore, we use the van Rossum distance, introduced in the preliminary elements (Section \ref{section:preliminaries}). Since this distance measures the similarity between \underline{two} activities, we extend its definition to function as follows:

    The goal is to determine if the pipeline can generate, based on a single sequence $\bm{x} \in P$, a population $P'$ whose overall activity is similar to that of $\bm{x}$. We calculate the average van Rossum distance between $\bm{x}$ and the other instances in the population $P$:
    $$d^P_\text{VR}(\bm{x}) = \dfrac{1}{|P| - 1}\sum_{\bm{x'}\in P, \bm{x} \ne \bm{x}'} d_\text{VR}(\bm{x}, \bm{x}'),$$
    and we calculate the average van Rossum distance between $\bm{x}$ and the other instances in $P'$, denoted as $d^{P'}_\text{VR}(\bm{x})$. For these calculations, we use the median ISI value of $\bm{x}$ as the value of $\tau$ in the van Rossum distance. This gives us a data-driven approach to adapt the hyperparameter of the distance, assuming that the time-scale that matters is the one of the baseline activity. Finally, we use the ratio of these distances as the evaluation metric:
    $$R_\text{VR}(\bm{x}) = \dfrac{d^{P'}_\text{VR}(\bm{x})}{d^{P}_\text{VR}(\bm{x})}.$$
    The interpretation is that if the pipeline is performing well, this ratio should be close to or less than 1. The denominator accounts for the inherent variability of the generation procedure. It is possible to obtain values of $R_\text{VR}(\bm{x}) < 1$ since the pipeline is conditioned solely on $\bm{x}$ and not on the entire population $P$.

    Additionally, we will study the robustness of the pipeline to noise in the input data. Specifically, we will investigate the impact of varying the intensity of input noise, $\sigma_\text{noise}$ on the DICs loss measured on the test set given the trained pipeline. This analysis will help us understand to what extent the model generalizes in the presence of noisy data, which is particularly important for laboratory use on data recorded \emph{in vitro}."""

    # group by g_u, g_s
    # loop over the groups

    def compute_van_rossum_distance(group):
        # Extract the first row as x
        x = group.iloc[0]
        # Extract the rest of the rows as y
        y = group.iloc[1:]
        # Compute the van Rossum distance
        for i in range(len(y)):
            y.iloc[i] = van_rossum_distance(x, y.iloc[i])
        # Compute the mean van Rossum distance
        mean_distance = y.mean()
        return mean_distance

    r_d = pd.DataFrame(columns=['g_u', 'g_s', 'denominator'])
    def process_group(args):
        name, group = args
        mean_distance = compute_van_rossum_distance(group['spiking_times'])
        return {'g_u': name[0], 'g_s': name[1], 'denominator': mean_distance}

    # Prepare arguments for multiprocessing
    groups = list(r.groupby(['g_u', 'g_s']))
    with Pool(processes=19) as pool:
        results = list(tqdm(pool.imap(process_group, groups), total=len(groups), desc="Processing groups"))

    # Convert results to DataFrame
    r_d = pd.DataFrame(results)

    # make a scatter plot of g_u, g_s, denominator as color
    plt.figure(figsize=(10, 10))
    plt.scatter(r_d['g_s'], r_d['g_u'], c=r_d['denominator'], cmap='viridis', s=5, alpha=0.5, vmax=1)
    plt.colorbar(label='Van Rossum distance')
    plt.xlabel('g_s')
    plt.ylabel('g_u')
    plt.title('Van Rossum distance between g_u and g_s')
    plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define path format and noise levels
path_format_noise = "tmp/cherry_picked_g_s_g_u_nl_{}.csv"
noise_levels = np.linspace(0, 5, 11)

# Initialize lists
mean_L_dics_noise = []
mean_mae_g_s_noise = []
mean_mae_g_u_noise = []
mean_sigma_s_noise = []
mean_sigma_u_noise = []

# Load values
for noise_level in noise_levels:
    path = path_format_noise.format(noise_level)
    cherry_picked_noise = pd.read_csv(path)

    mean_L_dics_noise.append(np.mean(cherry_picked_noise['L_dics']))
    mean_mae_g_s_noise.append(np.mean(np.abs(cherry_picked_noise['g_s'] - cherry_picked_noise['g_s_hat'])))
    mean_mae_g_u_noise.append(np.mean(np.abs(cherry_picked_noise['g_u'] - cherry_picked_noise['g_u_hat'])))
    mean_sigma_s_noise.append(np.mean(cherry_picked_noise['sigma_s']))
    mean_sigma_u_noise.append(np.mean(cherry_picked_noise['sigma_u']))

# Load bursting and spiking specific values
mean_mae_g_s_bursting_noise = []
mean_mae_g_u_bursting_noise = []
mean_mae_g_s_spiking_noise = []
mean_mae_g_u_spiking_noise = []

for noise_level in noise_levels:
    path = path_format_noise.format(noise_level)
    cherry_picked_noise = pd.read_csv(path)

    bursting = cherry_picked_noise[cherry_picked_noise["label"] == 1]
    spiking = cherry_picked_noise[cherry_picked_noise["label"] == 0]

    mean_mae_g_s_bursting_noise.append(np.mean(np.abs(bursting['g_s'] - bursting['g_s_hat'])))
    mean_mae_g_u_bursting_noise.append(np.mean(np.abs(bursting['g_u'] - bursting['g_u_hat'])))
    mean_mae_g_s_spiking_noise.append(np.mean(np.abs(spiking['g_s'] - spiking['g_s_hat'])))
    mean_mae_g_u_spiking_noise.append(np.mean(np.abs(spiking['g_u'] - spiking['g_u_hat'])))

# Helper function to compute percentage variation from baseline
def percent_variation_from_baseline(values):
    baseline = values[0]
    return [(v - baseline) / baseline * 100 for v in values]

# Compute percentage variations
L_dics_var = percent_variation_from_baseline(mean_L_dics_noise)
mae_g_s_var = percent_variation_from_baseline(mean_mae_g_s_noise)
mae_g_u_var = percent_variation_from_baseline(mean_mae_g_u_noise)
sigma_s_var = percent_variation_from_baseline(mean_sigma_s_noise)
sigma_u_var = percent_variation_from_baseline(mean_sigma_u_noise)

mae_g_s_bursting_var = percent_variation_from_baseline(mean_mae_g_s_bursting_noise)
mae_g_u_bursting_var = percent_variation_from_baseline(mean_mae_g_u_bursting_noise)
mae_g_s_spiking_var = percent_variation_from_baseline(mean_mae_g_s_spiking_noise)
mae_g_u_spiking_var = percent_variation_from_baseline(mean_mae_g_u_spiking_noise)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(noise_levels, L_dics_var, marker='o')
plt.xlabel('Noise level')
plt.ylabel('L_dics Variation (%)')
plt.title('Percentage variation in L_dics from no-noise baseline')
plt.grid()
plt.ylim(None, 25)
plt.xlim(0, 5)

plt.savefig("figures/percent_var_L_dics_noise.svg", format='svg', dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(noise_levels, mae_g_s_var, marker='o', label='g_s')
plt.plot(noise_levels, mae_g_u_var, marker='o', label='g_u')
plt.xlabel('Noise level')
plt.ylabel('MAE Variation (%)')
plt.title('Percentage variation in MAE (g_s, g_u) from no-noise baseline')
plt.legend()
plt.ylim(None, 25)
plt.xlim(0, 5)

plt.grid()
plt.savefig("figures/percent_var_MAE_g_s_g_u_noise.svg", format='svg', dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(noise_levels, sigma_s_var, marker='o', label='sigma_s')
plt.plot(noise_levels, sigma_u_var, marker='o', label='sigma_u')
plt.xlabel('Noise level')
plt.ylabel('Sigma Variation (%)')
plt.title('Percentage variation in sigma (s, u) from no-noise baseline')
plt.legend()
plt.ylim(None, 25)
plt.xlim(0, 5)
plt.grid()
plt.savefig("figures/percent_var_sigma_s_u_noise.svg", format='svg', dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(noise_levels, mae_g_s_bursting_var, marker='o', label='g_s (bursting)')
plt.plot(noise_levels, mae_g_s_spiking_var, marker='o', label='g_s (spiking)')
plt.xlabel('Noise level')
plt.ylabel('MAE g_s Variation (%)')
plt.title('Percentage variation in MAE g_s (bursting vs spiking)')
plt.legend()
plt.ylim(None, 25)
plt.xlim(0, 5)
plt.grid()
plt.savefig("figures/percent_var_MAE_g_s_bursting_spiking_noise.svg", format='svg', dpi=300)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(noise_levels, mae_g_u_bursting_var, marker='o', label='g_u (bursting)')
plt.plot(noise_levels, mae_g_u_spiking_var, marker='o', label='g_u (spiking)')
plt.xlabel('Noise level')
plt.ylabel('MAE g_u Variation (%)')
plt.title('Percentage variation in MAE g_u (bursting vs spiking)')
plt.legend()
plt.grid()
plt.ylim(None, 25)
plt.xlim(0, 5)
plt.savefig("figures/percent_var_MAE_g_u_bursting_spiking_noise.svg", format='svg', dpi=300)
plt.show()
```
